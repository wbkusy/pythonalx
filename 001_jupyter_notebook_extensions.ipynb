# Instalacja

## Jypyter - notebook

W docelowym środowisku instaluję jupyter:

    pip install jupyter

## Jupyter notebeook extensions

Instaluję rozszerzenia

    pip install jupyter_contrib_nbextensions

Instaluję pliki javascript i css

    jupyter contrib nbextension install --user
    
Po tej operacji konieczne może być zrestartowanie jupytera.


# Korzystanie:

## Uruchomienie

W konsoli systemowej wpisujemy:

    $> jupyter notebook
