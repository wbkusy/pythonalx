# Instalacja

## Jypyter - notebook

W docelowym rodowisku instaluj jupyter:

    pip install jupyter

## Jupyter notebeook extensions

Instaluj rozszerzenia

    pip install jupyter_contrib_nbextensions

Instaluj pliki javascript i css

    jupyter contrib nbextension install --user
    
Po tej operacji konieczne moe by zrestartowanie jupytera.
